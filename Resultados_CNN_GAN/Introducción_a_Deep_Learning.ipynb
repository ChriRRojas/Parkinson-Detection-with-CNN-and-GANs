## ¿Qué es el Deep Learning?

El Deep Learning, o aprendizaje profundo, es una rama del Machine Learning que utiliza redes neuronales de múltiples capas (o "profundas") para analizar datos complejos. Se basa en modelos de red neuronal que imitan el funcionamiento del cerebro humano para reconocer patrones en grandes cantidades de datos.

### Características principales:
- Permite el procesamiento de datos complejos como imágenes, audio y texto.
- Aprende automáticamente características importantes a medida que se entrena.
- Utiliza arquitecturas profundas para mejorar la precisión en tareas específicas.

En resumen, el Deep Learning ha revolucionado áreas como la visión por computadora, el procesamiento de lenguaje natural y la generación de imágenes.

## ¿Qué son las Redes Neuronales?

Una red neuronal es un modelo de aprendizaje inspirado en la estructura del cerebro humano. Está compuesta por capas de "neuronas" (o nodos) que están interconectadas. Cada neurona recibe información, la procesa y pasa el resultado a la siguiente capa.

### Componentes de una Red Neuronal:
- **Neuronas**: Cada nodo o neurona recibe una entrada, aplica una función (como la función sigmoide o ReLU) y genera una salida.
- **Capas**: La red neuronal está compuesta de capas:
  - **Capa de entrada**: Recibe los datos de entrada.
  - **Capas ocultas**: Realizan cálculos intermedios para aprender características complejas.
  - **Capa de salida**: Genera el resultado final de la red.

### Ejemplo de una Red Neuronal Sencilla
Vamos a crear una red neuronal básica usando TensorFlow y Keras:

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Definir el modelo
model = Sequential([
    Dense(16, activation='relu', input_shape=(10,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.summary()
```

> Explicación: Este código crea una red neuronal simple con dos capas ocultas y una capa de salida. Dense(16, activation='relu') significa que tenemos una capa con 16 neuronas y una función de activación ReLU.

## ¿Qué son las Redes Neuronales Convolucionales (CNN)?

Las redes neuronales convolucionales (CNN) son un tipo de red neuronal diseñada para procesar datos en forma de imágenes. Las CNN utilizan "capas convolucionales" que permiten extraer automáticamente características importantes de una imagen, como bordes, texturas, y formas.

### Estructura de una CNN
- **Capas Convolucionales**: Aplican filtros que destacan características visuales específicas.
- **Capas de Pooling**: Reducen la resolución espacial de las características, manteniendo la información importante.
- **Capas Densas**: Conectan las características extraídas a la capa de salida para hacer la predicción final.




image.png

### Ejemplo de una CNN básica en Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir el modelo CNN
model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_cnn.summary()
```

> Explicación: Aquí se define una CNN simple que puede ser usada para tareas de clasificación de imágenes. **Conv2D*(32, (3, 3), activation='relu')** crea una capa convolucional con 32 filtros de tamaño 3x3.


## ¿Qué es un Modelo GAN?

Un modelo GAN (Red Generativa Antagónica) es un tipo de modelo que se utiliza para generar datos sintéticos similares a los datos reales. Una GAN consta de dos redes neuronales:
- **Generador**: Crea datos sintéticos.
- **Discriminador**: Intenta distinguir entre datos reales y sintéticos.

El generador y el discriminador se entrenan juntos en un proceso adversarial, donde el generador intenta engañar al discriminador, y el discriminador intenta mejorar su habilidad para distinguir entre los datos reales y los generados.

### Ejemplo de una GAN en Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU

# Definir el Generador
generator = Sequential([
    Dense(128, input_dim=100),
    LeakyReLU(0.2),
    Dense(784, activation='tanh')
])

# Definir el Discriminador
discriminator = Sequential([
    Dense(128, input_dim=784),
    LeakyReLU(0.2),
    Dense(1, activation='sigmoid')
])

# Compilar el Discriminador
discriminator.compile(optimizer='adam', loss='binary_crossentropy')

discriminator.summary()
generator.summary()
```

> Explicación: Este ejemplo muestra cómo crear un generador y un discriminador simples. El generador toma un vector aleatorio y genera datos sintéticos, mientras que el discriminador intenta clasificar los datos como reales o falsos.
