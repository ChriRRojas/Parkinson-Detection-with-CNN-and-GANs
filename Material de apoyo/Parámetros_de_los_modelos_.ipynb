### Parámetros del Entrenamiento
Estos son los parámetros utilizados para entrenar la **GAN** y la **CNN**:

#### 1. **Generador y Discriminador (GAN)**
- **Batch Size**: Número de imágenes procesadas en cada paso de entrenamiento (por ejemplo, `16`).
- **Épocas**: Número de veces que se recorre el dataset completo (`100` épocas para la GAN).
- **Dimensión del Ruido (`dim_ruido`)**: Dimensión del vector de ruido para el generador (`100`).

#### 2. **Modelo CNN**
- **Épocas**: Número de veces que se recorrerá el dataset completo (`15` épocas).
- **Batch Size**:
  - **Entrenamiento**: `16`
  - **Validación**: `32`
- **Optimizador**: Adam con tasa de aprendizaje `1e-4`.
- **Función de Pérdida**: `binary_crossentropy` para clasificación binaria (Parkinson o Control).

### Resumen de los Modelos
- **Modelo GAN**:
  - **Generador** crea imágenes a partir de ruido con **1,235,392** parámetros entrenables.
  - **Discriminador** clasifica imágenes como reales o falsas, con **242,561** parámetros entrenables.
- **Modelo CNN**:
  - Se utiliza para clasificar entre imágenes de **Parkinson** y **control**, con **3,303,769** parámetros entrenables.

Cada modelo tiene un rol específico: **el generador** intenta crear imágenes que parezcan reales, **el discriminador** intenta diferenciar entre reales y falsas, y **la CNN** se entrena para clasificar las imágenes entre las dos clases objetivo, usando tanto datos reales como imágenes generadas.
